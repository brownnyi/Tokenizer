```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras import backend as K

class WordAttentionLayer(layers.Layer):
    def __init__(self, **kwargs):
        super(WordAttentionLayer, self).__init__(**kwargs)

    def build(self, input_shape):
        self.W_word = self.add_weight(name="W_word", shape=(input_shape[-1], 1), initializer="normal")
        self.b_word = self.add_weight(name="b_word", shape=(input_shape[1], 1), initializer="zeros")
        super(WordAttentionLayer, self).build(input_shape)

    def call(self, x):
        e = K.tanh(K.dot(x, self.W_word) + self.b_word)
        a = K.softmax(e, axis=1)
        output = x * a
        return K.sum(output, axis=1)

    def compute_output_shape(self, input_shape):
        return input_shape[0], input_shape[-1]

class SentenceAttentionLayer(layers.Layer):
    def __init__(self, **kwargs):
        super(SentenceAttentionLayer, self).__init__(**kwargs)

    def build(self, input_shape):
        self.W_sentence = self.add_weight(name="W_sentence", shape=(input_shape[-1], 1), initializer="normal")
        self.b_sentence = self.add_weight(name="b_sentence", shape=(input_shape[1], 1), initializer="zeros")
        super(SentenceAttentionLayer, self).build(input_shape)

    def call(self, x):
        e = K.tanh(K.dot(x, self.W_sentence) + self.b_sentence)
        a = K.softmax(e, axis=1)
        output = x * a
        return K.sum(output, axis=1)

    def compute_output_shape(self, input_shape):
        return input_shape[0], input_shape[-1]

def create_hand_model(sentence_length=16, word_length=64, num_classes=2):
    input_layer = layers.Input(shape=(sentence_length, word_length))

    # Word Encoder
    word_encoder = layers.TimeDistributed(layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=word_length))(input_layer)
    word_encoder = layers.TimeDistributed(layers.GRU(50, return_sequences=True))(word_encoder)
    word_attention = WordAttentionLayer()(word_encoder)

    # Sentence Encoder
    sentence_encoder = layers.GRU(100, return_sequences=True)(word_encoder)
    sentence_attention = SentenceAttentionLayer()(sentence_encoder)

    # Fully Connected Layer
    fc_layer = layers.Dense(50, activation='relu')(sentence_attention)
    output_layer = layers.Dense(num_classes, activation='softmax')(fc_layer)

    model = models.Model(inputs=input_layer, outputs=output_layer)

    return model

# 모델 생성
model = create_hand_model()

# 모델 컴파일
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 모델 요약
model.summary()
```
